In this session we will study, by means of computer simulations in PyDSTool and continuation analysis using AUTO (within PyDSTool), the bifurcation behavior of a gene regulation circuit described by
ordinary differential equations.

Following up practical \#3, we consider again the phosphorylation/dephosphorylation cycle shown in the figure below.
In this circuit, phosphorylation is produced by both an stimulus signal $S$ and by the phosphorylated
protein itself, in the form of a positive feedback loop. Assuming that the latter process is cooperative, the model
of this system is:

<img style="margin: 20px; float: right" src="http://dsb.upf.edu/biocomp/figs/irrevers.png" width="15%"/>

$$
\frac{dA_p}{dt}=k_p SA+\beta\frac{A_p^n}{K^n+A_p^n}-k_dA_p
$$

where $k_p$ is the activation (phosphorylation) rate, $k_d$ is the inactivation (dephosphorylation) rate, $A$ is the
concentration of the unphosphorylated protein, and $A_p$ is the concentration of the phosphorylated
protein. 

The total protein concentration, $A_T=A+A_p$, is assumed to be constant, and the parameter values
will be $K=50$, $A_T=100$, $k_d=0.13$, $\beta=10$, $k_p=0.1$, and $n=4$.

First, import all necessary packages.

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import PyDSTool as dst
```

After the initial imports (notice we also imported the `PyDSTool` package) we need to define our dynamical system in order to integrate it. We will need to write the equations as strings (in a `C`-compatible syntaxis), the initial conditions and the parameter values. For all this we will use Python dictionaries.

```python
# we must give a name
DS_arguments = dst.args(name='Irrev_stimulation')

# Define the equations
DS_arguments.varspecs = {'Ap':'kp*S*(At-Ap)+b*pow(Ap,n)/(pow(K,n)+pow(Ap,n))-kd*Ap'}

# initial conditions
DS_arguments.ics      = {'Ap': 0}

# parameters
DS_arguments.pars = {'S' : 0.5,
                     'kp' : 0.1,
                     'kd' : 0.13,
                     'At' : 100,
                     'b' : 10,
                     'K' : 50,
                     'n' : 4,}

# Set the time domain for the integration
DS_arguments.tdomain = [0,200]

# Generate the ODE system as ode
ode  = dst.Generator.Vode_ODEsystem(DS_arguments)
```

The system can be then integrated from the `ode` object.

```python
traj = ode.compute('irreversible')   # integrate ODE
pts  = traj.sample(dt=0.1)           # extract one point every dt
```

Now, `pts` contains one entry with all the time points (`pts['t']`) and one entry for every variable of our system (in this case `pts['Ap']`). This can easily be used to plot the time evolution of the system.

```python
# PyPlot commands
plt.figure(figsize=(12,6))
plt.plot(pts['t'], pts['Ap'], color='r')
plt.xlabel('time',fontsize=20)                  # Axes labels
plt.ylabel('concentration',fontsize=20)         # ...
plt.ylim([0,100])                   # Range of the y axis
plt.title('time trace',fontsize=20)              # Figure title from model name
plt.tick_params(labelsize=15)
plt.show()
```

Now, we want to study the behaviour of the system for different values of $S$. Taking into account that the value of a parameter in `ode` can be changed by

    ode.set(pars={ PARAMETER_NAME: PARAMETER_VALUE})

you are asked to do the integration of the system with different values of $S$ ranging from 0 and 1 (use a `for` loop).
Take into account that you do not need to define the system again, just change the parameter, integrate, sample the result and plot it.
What happens in response to these variations in $S$? If you detect a sudden change in the equilibrium state of the system for varying $S$, indicate the value of $S$ for which the jump happens.

```python
plt.figure(figsize=(12,6))

for S_value in np.linspace(0, 1 , 7):
    ode.set( pars = {'S' : S_value})
    
    traj_equil = ode.compute('irreversible')
    pts = traj_equil.sample(dt=0.1)
    
    plt.plot(pts['t'], pts['Ap'], label="S = %g" %S_value)

plt.xlabel('time',fontsize=20)                  # Axes labels
plt.ylabel('concentration',fontsize=20)         # ...
plt.ylim([0,100])                   # Range of the y axis
plt.title('increasing S - starting from 0',fontsize=20)              # Figure title from model name
plt.legend()
plt.tick_params(labelsize=15)
plt.show()
    
```

To make sense of the behavior observed so far, we will now perform a _continuation_ of
the equilibrium states of the system. This numerical method consists on, starting from a known
equilibrium point, varying slowly one parameter and tracking how the equilibrium state and
its stability change with that parameter. `PyDSTool` has a build-in interface to interact with [AUTO](http://cmvl.cs.concordia.ca/auto/}{http://cmvl.cs.concordia.ca/auto/)
to perform this task. AUTO is an extremely popular continuation software that was developed independently, and which PyDSTool makes use of.

Set $k_d$ back to 0.13 and $S=0$. Run again the simulation making sure it reaches a steady state. Since we need to start the continuation from the equilibrium steady state, set as the initial conditions the steady state obtained.

```python
ode.set(ics = {'Ap': 0})
ode.set(pars = {'S': 0})

ode.set(pars = {'kd': 0.13})

traj = ode.compute('irreversible')
pts  = traj.sample(dt=0.1)  

ode.set(ics = {'Ap': pts['Ap'][-1]})
```

Then we need to set up the configuration parameteres for the continuation. We will use pretty much the default ones. The parameters will be stored in the `Continuation_args` object that you can use in the rest of the notebook.

```python
# 'EP-C' stands for Equilibrium Point Curve. The branch will be labeled 'EQ1'.
Continuation_args = dst.args(name='EQ1', type='EP-C')
# control parameter(s) (it should be among those specified in DSargs.pars)
Continuation_args.freepars     = ['S']          
Continuation_args.MaxNumPoints = 950 
Continuation_args.LocBifPoints = 'all'                 # detect limit points / saddle-node bifurcations
Continuation_args.SaveEigen    = True                  # to tell unstable from stable branches
Continuation_args.StopAtPoints = ['B']
Continuation_args.force        = True
```

Now, we need to specify the upper and lower limit of $S$ that we will allow during our exploration, set up the continuation object, define a new curve and then do the actual computation with the `forward` method. Finally, the `display` method should allow us to plot the continuation diagram specifying the axis.

The bifurcation diagram should appear, with the lines of stable equilibrium points (stable branch) represented as solid lines, and an unstable branch as dashed. Compare the values represented in the stable branches for varying levels of $S$, with the simulations made in sections above.

Note that bifurcations are also marked in the plot. The most common bifurcation points that AUTO detects are LP (limit point, saddle-node bifurcation), BP (branch point, transcritical bifurcation), PD (period-doubling bifurcation), HB (Hopf bifurcation), TR (torus bifurcation). Identify the type of the two bifurcation points in this case.

```python
ode.parameterDomains['S'].set([-0.5,1])  # Define the range within we want to explore the variation of S

PC = dst.ContClass(ode)            # Set up continuation class
PC.newCurve(Continuation_args)

PC['EQ1'].forward()

plt.figure(figsize=(12,6))
PC.display(['S','Ap'], stability=True, figure='fig1')
plt.xlim(0,1)
```

Repeat the continuation above for $k_d=0.11$. To get to it, choose the initial condition for $S=0$ such that the system falls onto the upper branch, and start the continuation from there. Set the limits of the x axis of the final plot from $0$ to $1$.

```python
ode.set(ics = {'Ap': 0})
ode.set(pars = {'S': 0})

ode.set(pars = {'kd': 0.11})

traj = ode.compute('irreversible')              # integrate ODE
pts  = traj.sample(dt=0.1)  

ode.set(ics = {'Ap': pts['Ap'][-1]})

ode.parameterDomains['S'].set([-0.5,1])
PC = dst.ContClass(ode)            # Set up continuation class
PC.newCurve(Continuation_args)

PC['EQ1'].forward()

plt.figure(figsize=(12,6))
PC.display(['S','Ap'], stability=True, figure='fig1')
plt.xlim(0,1)
```
